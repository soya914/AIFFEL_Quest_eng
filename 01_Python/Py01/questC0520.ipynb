# Q1-2 회문 분류기

```python
def is_palindrome(word):
    return word == word[::-1]


def palindrome_classifier(targets):

    result = {
        '회문O': [],
        '회문X': []
    }

    for word in targets:

        if is_palindrome(word):
            result['회문O'].append(word)

        else:
            result['회문X'].append(word)


    palindrome_count = int(len(result['회문O']))

    palindrome_ratio = float(
        round(
            palindrome_count / len(targets),1
        )
    )

    return result, palindrome_count, palindrome_ratio



targets = [
'기러기',
'까마귀',
'토마토',
'스윗소스',
'스위스',
'인도인',
'한국인',
'별똥별',
'우영우',
'우향우'
]

print(palindrome_classifier(targets))   
```

# Q2-2 성적 분석기

```python
scores = {

    '유원영':{
        '국어':94,
        '영어':91,
        '수학':89
    },

    '설지민':{
        '국어':88,
        '영어':96,
        '수학':90
    },

    '신윤아':{
        '국어':86,
        '영어':88,
        '수학':77
    },

    '장윤아':{
        '국어':98,
        '영어':76,
        '수학':92
    }
} 
def score_analysis(name):

    student = scores[name]


    korean_avg = (
        scores['유원영']['국어']
        + scores['설지민']['국어']
        + scores['신윤아']['국어']
        + scores['장윤아']['국어']
    ) / 4


    english_avg = (
        scores['유원영']['영어']
        + scores['설지민']['영어']
        + scores['신윤아']['영어']
        + scores['장윤아']['영어']
    ) / 4


    math_avg = (
        scores['유원영']['수학']
        + scores['설지민']['수학']
        + scores['신윤아']['수학']
        + scores['장윤아']['수학']
    ) / 4


    student_avg = (
        student['국어']
        + student['영어']
        + student['수학']
    ) / 3


    class_avg = (
        korean_avg
        + english_avg
        + math_avg
    ) / 3


    print(f'[{name} 성적 분석]')

    print(
        f'국어: {student["국어"]}점, '
        f'학급 평균({korean_avg:.2f}) 대비 '
        f'{student["국어"]-korean_avg:+.2f}'
    )

    print(
        f'영어: {student["영어"]}점, '
        f'학급 평균({english_avg:.2f}) 대비 '
        f'{student["영어"]-english_avg:+.2f}'
    )

    print(
        f'수학: {student["수학"]}점, '
        f'학급 평균({math_avg:.2f}) 대비 '
        f'{student["수학"]-math_avg:+.2f}'
    )


    print()

    print(
        f'전과목 평균:{student_avg:.2f}점, '
        f'학급 평균({class_avg:.2f}) 대비 '
        f'{student_avg-class_avg:+.2f}'
    )

score_analysis('유원영')
score_analysis('설지민')
score_analysis('신윤아')
score_analysis('장윤아')
```
