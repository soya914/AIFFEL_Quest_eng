# 파이썬을 사용하여 간단한 계산기 프로그램 만들기

```python
from math import pow


# 유효한 정수 확인 함수
def valid_number(input_text):
  # 정수가 아닌 경우 ValueError를 처리하여 적절한 오류 메시지를 출력 -> except
  # 정수가 입력될 때 까지 “잘못된 입력입니다. 정수를 입력해주세요.”를 출력하며 -> print
  # 다시 입력받기를 시도 -> while True

  while True: # 완료될 때까지 무한반복
    try:
      num = int(input(input_text))

    # 정수가 아닌 경우 오류 메시지 출력
    except ValueError:
      print("잘못된 입력입니다. 정수를 입력해주세요.")
    else:
      return num


# 유효한 연산자 확인 함수
def valid_operator():

  while True: # 완료될 때까지 무한반복
    # 조건으로 지정된 연산자 리스트

    valid_operator = ["+", "-", "*", "/", "**"]
    operator = input("연산자를 입력하세요 (+,-,*,/,**): ")


    # valid_operator에 들어있는 연산자가 operator로 입력되었다면 operator를 반환
    if operator in valid_operator:
      return operator

    # operator로 지정된 연산자가 아닌 입력이 들어올 경우
    else:
      print("지원하지 않는 연산자입니다.")



# 계산 진행 함수
def calculation(num1, num2, operator):

  if operator == "+":
    result = num1 + num2

  elif operator == "-":
    result = num1 - num2

  elif operator == "*":
    result = num1 * num2

  elif operator == "/":
    try:
      result = num1 / num2
    except ZeroDivisionError:
      result = "0으로 나눌 수 없습니다."

  # math 모듈로 import한 제곱 함수 사용
  elif operator == "**":
    result = pow(num1, num2)

  return result

# 계산기 종료 확인 함수
# True를 반환하면 계산기 종료
def choice_end():
  while True:
    choice = input("계산기를 계속 사용하시겠습니까? (y/n): ")

    # 계산기를 종료할지, 재실행할지 선택
    if choice.lower() == "n":
      return True
    elif choice.lower() == "y":
      return False
    else:
      print("잘못된 입력입니다. 다시 입력해주세요.")


# 계산기 메인 함수
def calculator_app():

  print(f"{'= '*10}계산기를 시작합니다.{' ='*10}")

  while True:

    # 정수를 입력 받고 유효성 확인 후 정수 반환
    num1 = valid_number("첫 번째 정수를 입력하세요: ")
    num2 = valid_number("두 번째 정수를 입력하세요: ")

    # 연산자를 입력 받고 유효성 확인 후 연산자 operator 반환
    operator = valid_operator()

    # 두 정수와 연산자를 계산 함수를 호출하여 연산하고 결과값 result 반환
    result = calculation(num1, num2, operator)

    # 반환된 정수, 연산자, 결과값 출력
    print(f"{num1} {operator} {num2} = {result}")

    # 계산기를 통해 계속 계산할 것인지 입력 받고 확인 후 bool 반환
    # choice_end() == True 면 계산기 종료, False면 계산기 재진행
    if choice_end():
      break


  print(f"{'= '*10}계산기를 종료합니다.{' ='*10}")



### 실행
calculator_app() 
```

# 디버깅 내용

ValueError, ZeroDivisionError, 잘못된 연산자 입력, 종료 선택 오류를 확인하고 수정하였다.
예외처리를 사용하여 잘못된 입력이 들어와도 프로그램이 바로 종료되지 않고, 사용자에게 안내 메시지를 출력한 뒤 다시 입력받을 수 있도록 만들었다.
