# 직사각형, 평행사변형, 사다리꼴 넓이 구하는 매서드 만들기

```python
# 클래스 만들기 
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
        
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    #직사각형 매서드
    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    #평행사변형 매서드
    def par(self):
        base, height = map(int,input('밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>>').split(','))
        area = base * height
        result = '평행사변형의 넓이는 : ' + str(area)
        return result

    #사다리꼴 매서드
    def trape(self):
        top, bottom, height = map(int,input('윗변, 아랫변, 높이를 입력하세요. 예시 : 윗변,아랫변,높이\n >>>').split(','))
        area = (top + bottom) * height / 2
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result


#객체 만들기
a= Square()

#넓이 구하기
if a.square == 1:
    print(a.rect())

elif a.square == 2:
    print(a.par())

elif a.square == 3:
    print(a.trape())
```

# Kiosk 클래스 생성하고 주문, 추가주문 매서드 작성하기

```python
## 클래스 생성하기
class Kiosk:
    def __init__(self):
        self.menu = ['americano','latte','mocha','yuza_tea','green_tea','choco_latte']
        self.price = [2000,3000,3000,2500,2500,3000]


    # 메뉴 출력 매서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1,self.menu[i],':',self.price[i],'원')


    ## 주문 메서드
    def menu_select(self):

        self.order_menu = []
        self.order_price = []
        self.price_sum = 0
        
        n= int(input("음료를 번호를 입력하세요 : "))
        self.price_sum = self.price[n-1]

            
        # 음료 온도 물어보기
        t = 0
        while t != 1 and t != 2:
            t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))

            # 온도 지정
            if t == 1:
                self.temp = "HOT"
            elif t == 2:
                self.temp = "ICE"
            else:
                print("1과 2 중 하나를 입력하세요.\n")

         #주문하기
        self.order_price.append(self.price[n-1])
        self.order_menu.append(self.temp + ' ' + self.menu[n-1])
        print('주문 음료', self.temp, self.menu[n-1], ':', self.price[n-1], '원')

    #추가 주문 또는 지불하기
        while n != 0:
            0

            0
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))
            
            if n > 0 and n < len(self.menu) + 1:


            #추가 음료온도
                t = 0
                while t != 1 and t != 2:

                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    
                    if t == 1:
                        self.temp = "HOT"
                    elif t == 2:
                        self.temp = "ICE"
                    else:
                        print("다시 입력하세요.\n")

                #가격 추가
                self.order_price.append(self.price[n-1])
                
                # 메뉴 추가
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])
            
                # 총 금액 누적
                self.price_sum += self.price[n-1]
                 
                print('주문 음료',self.temp,self.menu[n-1],':',self.price[n-1],'원\n','합계 :',self.price_sum,'원')
            
        # 지불하기
            elif n == 0:
                print("주문이 완료되었습니다.")
                print(self.order_menu,self.order_price)
            else:
                print("없는 메뉴입니다.")

    # 지불
    def pay(self):

    # 합계 출력
        print(f"총 합계 금액은 {self.price_sum}원 입니다.")

        self.pay_method = 0

        while self.pay_method != 1 and self.pay_method != 2:

            self.pay_method = int(input("현금 결제는 1번, 카드 결제는 2번을 입력해주세요. : "))

            if self.pay_method == 1:
                print("직원을 호출하겠습니다.")

            elif self.pay_method == 2:
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")

            else:
                print("다시 결제를 시도해 주세요.")

    # 주문표 메서드
    def table(self):

        # 외곽
        outline1 = "-" + "-" * 30 + "-"
        outline2 = "|" + " " * 31 + "|"

        # 외곽 출력
        print(outline1)

        for i in range(5):
            print(outline2)

        # 주문 상품 + 해당 금액
        for i in range(len(self.order_menu)):

            print(self.order_menu[i],':',self.order_price[i])

        # 합계 금액
        print('합계 금액 :', self.price_sum)

        # 외곽 출력
        for i in range(5):
            print(outline2)

        print(outline1)

        
#주문하기
a = Kiosk()
a.menu_print()
a.menu_select()
a.pay()
a.table()
```
