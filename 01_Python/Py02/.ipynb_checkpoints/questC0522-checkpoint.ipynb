## Q1.클로저를 활용한 문제 해결하기

```python
def find_min_max(numbers):

    # 최솟값, 최댓값 초기화
    min_value = numbers[0]
    max_value = numbers[0]

    
    # 리스트를 순회하며 최솟값과 최댓값 갱신
    for num in numbers:
        if num < min_value:
            min_value = num
        if num > max_value:
            max_value = num  

    def find_min():
        # find_min_max 종료 후에도 min_value를 클로저로 참조
        return min_value

    def find_max():
        # find_min_max 종료 후에도 max_value를 클로저로 참조
        return max_value

    # 두 클로저 함수를 튜플로 반환  
    return find_min, find_max

# 주어진 숫자 리스트와 함수 호출
numbers = [10, 5, 8, 12, 3, 7]
find_min, find_max = find_min_max(numbers)

# 결과확인 출력
print("최솟값:", find_min())
print("최댓값:", find_max())
```

# Q2. 클로저를 활용한 문제 해결하기 

```python
def counter(func):
    # 실행 횟수를 기억할 지역 변수 선언
    count = 0

    # 클로저를 활용한 데코레이터의 내부 함수
    def inner_func(*args, **kwargs):
        nonlocal count # 지역 변수인 count를 수정하기 위해 nonlocal로 선언
        count += 1
        
        # 원래의 함수를 실행하여 "Hello Aiffel!"를 출력  
        result = func(*args, **kwargs)

        # 함수 이름과 실행횟수를 출력
        print(f"{func.__name__} 실행횟수: {count}")
        
        return result
        
    return inner_func

        
# 데코레이터 적용
@counter
def say_hello ():
    print("Hello Aiffel!")

#5번 반복실행
for i in range(5):
    say_hello()
```
